```python
# importing random module
import random
```

```python
# user defined function update the moves of checker
def generate(val,num):
    snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}
    ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}
    s = val+num
    k=s
    if s<100:
        for i in snakes.keys():
            if s == i:
                k=snakes[i]
                print("Sorry for thisssss.....")
        for i in ladders.keys():
            if s== i:
                k=ladders[i]
                print("Yeahhhh!!!!! Ladder.......")
        print("You got a ",num)
        print("Your Final Position is ",k)
        return k
    elif s>100:
        print("You got a ",num)
        print("Your Final Position is ",val)
        return val
    elif s==100:
        return s
```

```python
#user defined function to check if  the number entered manually if a home for snake or is a ladder for the player to get close to win
def check(val):
    snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}
    ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}
    k=0
    for i in snakes.keys():
            if val == i:
                k=snakes[i]
                print("Sorry for thisssss.....")
    for i in ladders.keys():
            if val== i:
                k=ladders[i]
                print("Yeahhhh!!!!! Ladder.......")
    if k!=0:
        return k
    else:
        return val
```

```python
print("---------Welcome to Snakes & Ladders Game--------")
print("Rules of the game:\n to roll the dice players can type roll\n players can also enter any number between 1-20\n to exit the game enter quit\n")
#user input of nameof players
player1 = input("Enter the name of Player1: ")
player2 = input("Enter the name of Player2: ")
print("---------Let's start the Game--------")
counter1 = 0
counter2 = 0
c =0
while 1:
    inp1=input("{}: ".format(player1)).lower()
    if inp1 == 'roll':
        num = random.randint(1,6)
        k1 = generate(counter1,num)#function call
        counter1 = k1
        if k1 == 100:
            print(f"{player1} Won!!!")
            c =c +1
            break
    elif inp1 == 'quit':
        print(f"{player2} Won!!!")
        break
    elif inp1.isdigit():#using .isdigit() to check if the string contains a number
        if int(inp1)>=1 and int(inp1)<=20:
            counter1 = counter1 + int(inp1)
            counter1 = check(counter1)#function call
            if counter1<100:
                print("Your Final Position is ",counter1)
            elif counter1 == 100:
                print(f"{player1} Won!!!")
                c = c+1
                break
            elif counter1>100:
                counter1 = counter1 - int(inp1)
                print("Your Final Position is ",counter1)
            else:
                print("Wrong Input!!!")#if inp2 is less than 1 and greater than 20 then this will be executed
                break
                
    else:
        print("Wrong Input---Restart the game------")
        break
        
    inp2=input("{}: ".format(player2)).lower()
    if inp2== 'roll':
        num = random.randint(1,6)
        k2 = generate(counter2,num)#function call
        counter2=k2
        if k2 == 100:
            print(f"{player2} Won!!!")
            c = c+1
            break
    elif inp2== 'quit':
        print(f"{player1} Won!!!")
        break
    elif inp2.isdigit():#using .isdigit() to check if the string contains a number
        if int(inp2)>=1 and int(inp2)<=20:
            counter2 = counter2 + int(inp2)
            counter2 = check(counter2)#function call
            if counter2<100:
                print("Your Final Position is ",counter2)
            elif counter2 == 100:
                print(f"{player2} Won!!!")
                c = c+1
                break
            elif counter2>100:
                counter2 = counter2 - int(inp2)
                print("Your Final Position is ",counter2)
        else:
                print("Wrong Input!!!")#if inp2 is less than 1 and greater than 20 then this will be executed
                break
    else:
        print("Wrong Input---Restart the game------")
        break
if c > 0:
    print("-------Game Successfully Finished--------")
else:
    print("Thanks!!!")
```
